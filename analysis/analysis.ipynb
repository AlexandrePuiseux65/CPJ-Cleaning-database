* *Verification Barriers*: If information is insufficient or impossible to verify, a case may remain *unconfirmed* indefinitely.
* *Data Labeling*: Because data comes from diverse global sources, inconsistencies or errors in labeling can occur.
* *Intentional Gaps*: Specific data points may be listed as *unknown* (such as place of death) intentionally when information cannot be verified, rather than being a mistake.

```python
# Temporarily normalize text to detect inconsistent labels
# (e.g. differences in capitalization or spacing)
def normalize_text(series):
    return (
        series.astype("string")
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", " ", regex=True)
    )

# Identify categorical values that differ only by formatting
# by comparing original values with their normalized version
def find_case_variants(df, col):
    temp = pd.DataFrame({"original": df[col].dropna(), "normalized": normalize_text(df[col].dropna())})
    grouped = (temp.groupby("normalized")["original"].unique().reset_index())

    grouped["n_variants"] = grouped["original"].apply(len)
    variants = grouped[grouped["n_variants"] > 1]
    return variants.sort_values("n_variants", ascending=False)
```

Now, we also want to examine whether different columns contain overlapping values.
The goal is not only to identify similarities between variables, but also to detect possible data quality issues, such as values that may have been entered in the wrong column.

To do so, we compute a similarity matrix between categorical columns using the Jaccard similarity coefficient:

$$Similarity = \frac{|A \cap B|}{|A \cup B|}$$

For each pair of columns, we compare their sets of unique values.
A high similarity score means that the two columns share many values. This can reveal:

1. redundancy between columns
2. conceptual overlap
3. or values that may be misplaced

After identifying suspicious column pairs, we can list the shared values and inspect them manually in order to confirm whether they are valid or incorrect.
For example, if two semantically different columns share the same values, those shared elements can be extracted and reviewed individually to determine whether they correspond to valid overlap or to a placement error.

```python
# Comparison of element between to of the same colunm, by using Jaccard similarity
# Import seaborn and matplotlibs libraries
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

columns = df.select_dtypes(include="object").columns
matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)

# Compute pairwise Jaccard similarity between categorical columns
for col1 in columns:
    # Create a set of unique non-null values for the first column
    set1 = set(df[col1].dropna().astype(str))

    for col2 in columns:
        # Create a set of unique non-null values for the second column
        set2 = set(df[col2].dropna().astype(str))

        # Compute intersection and union of both sets
        intersection = len(set1 & set2)
        union = len(set1 | set2)

        # Calculate Jaccard similarity score
        if union == 0:
            score = 0
        else:
            score = intersection / union

        # Store the similarity score in the matrix
        matrix.loc[col1, col2] = score
```

```python
# Visualize the similarity matrix using a heatmap
plt.figure(figsize=(24,10))

# Hide cells with zero similarity to improve readability
mask = (matrix == 0)
sns.heatmap(
    matrix,
    mask=mask,
    cmap="YlOrBr",
    annot=True,
    fmt=".3%",
    vmax=0.1,
    norm=LogNorm(vmin=0.0001, vmax=1),
    linewidths=1,
    cbar=False
)

plt.title("Column Similarity Matrix", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

## II. Data Cleaning & Preparation

As identified in the previous section, the dataset contains several data quality issues that prevent it from being directly used for analysis and visualisation. Therefore, a structured cleaning and preparation process is required in order to ensure consistency, improve data reliability, and extract meaningful information.

The dataset is compiled by the Committee to Protect Journalists through investigative research based on multiple sources. Consequently, some variables contain missing values, inconsistent formatting, or multiple categories within the same field due to differences in the original reporting sources.

The cleaning process is organised into several steps:

a. Standardisation of categorical labels  
b. Correction of inconsistencies across columns  
c. Handling of multi-value categorical fields  
d. Removal of redundant or unused variables  
e. Creation of derived relational tables  

Regarding the **organisation** variable, no modification is applied. Organisation names often contain contextual information such as geographic identifiers or extended institutional titles. As these values represent full entity names rather than categorical labels, standardisation or splitting could introduce inaccuracies.

```python
# For columns containing a single value per cell
def normalize_single_value(series, corrections):
    return series.apply(
        lambda x: corrections.get(str(x).strip().lower(), x) if pd.notna(x) else x
    )

# For columns containing multiple comma-separated values
def normalize_multi_value(series, corrections, sep=","):
    def clean_cell(x):
        if pd.isna(x):
            return x
        
        parts = [part.strip() for part in str(x).split(sep)]
        corrected_parts = [corrections.get(part.lower(), part) for part in parts]
        return ", ".join(corrected_parts)
    
    return series.apply(clean_cell)
```

```python
# -----------------------------
# Single-value columns
# -----------------------------
df_clean["employedAs"] = normalize_single_value(df_clean["employedAs"], employedAs_corrections)
df_clean["organizations"] = normalize_single_value(df_clean["organizations"], organizations_corrections)

df_clean["location"] = normalize_single_value(df_clean["location"], location_corrections)
df_clean["locality"] = normalize_single_value(df_clean["locality"], location_corrections)

df_clean["lengthOfSentence"] = normalize_single_value(df_clean["lengthOfSentence"], length_sentence_corrections)
df_clean["sentence"] = normalize_single_value(df_clean["sentence"], sentence_corrections)

# -----------------------------
# Multi-value columns
# -----------------------------
df_clean["jobs"] = normalize_multi_value(df_clean["jobs"], jobs_corrections)
df_clean["coverage"] = normalize_multi_value(df_clean["coverage"], coverage_corrections)
df_clean["mediums"] = normalize_multi_value(df_clean["mediums"], mediums_corrections)
df_clean["charges"] = normalize_multi_value(df_clean["charges"], charges_corrections)
```

```python
def clean_location_row_simple(row):
    loc = row["location"]
    reg = row["region"]

    if pd.isna(loc):
        return pd.Series([loc, reg])

    loc = str(loc).strip().rstrip(",").replace("Xi‚Äôan", "Xi'an")

    if "," not in loc:
        return pd.Series([loc, reg])

    left, right = [p.strip() for p in loc.split(",", 1)]

    if pd.isna(reg):
        reg = right

    return pd.Series([left, reg])

df_clean[["location", "region"]] = df_clean.apply(clean_location_row_simple, axis=1)
```

## <u>C) Global validation</u>

```python
columns = df_clean.select_dtypes(include="object").columns
matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)

for col1 in columns:
    set1 = set(df_clean[col1].dropna().astype(str))

    for col2 in columns:
        set2 = set(df_clean[col2].dropna().astype(str))

        intersection = len(set1 & set2)
        union = len(set1 | set2)

        if union == 0:
            score = 0
        else:
            score = intersection / union

        matrix.loc[col1, col2] = score

plt.figure(figsize=(24,10))

mask = (matrix == 0)

sns.heatmap(
    matrix,
    mask=mask,
    cmap="YlOrBr",
    annot=True,
    fmt=".3%",
    vmax=0.1,
    norm=LogNorm(vmin=0.0001, vmax=1),
    linewidths=1,
    cbar=False
)

plt.title("Column Similarity Matrix", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```
